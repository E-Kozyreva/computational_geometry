### Import libs

```python
import time
from random import randint, uniform
import matplotlib.pyplot as plt

from algorithms.graham_scan import GrahamScan
import algorithms.jarvis_march as jarvis
```

### Алгоритм Грэхема (Graham scan)

```python
points = [(randint(25, 175), randint(25, 175)) for _ in range(50)]
g = GrahamScan(points)
```

### Output

```python
print('Points:', end=' ')
for p in range(len(points)):
    if p % 10 == 0:
        print()
    print(points[p], end=' ')

print(f"\n\nConvex hull: {g.convex_hull}")
g.draw()
```
